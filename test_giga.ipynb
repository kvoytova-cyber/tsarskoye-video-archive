```python
from gigachat import GigaChat
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
api_key = os.getenv("GIGA_KEY")
```

```python
giga = GigaChat(
   credentials=api_key,
   verify_ssl_certs=False
)
```

```python
response = giga.chat("Привет! Как дела?")
print(response.choices[0].message.content)
```
